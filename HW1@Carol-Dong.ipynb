### Question 1

1. a) *H<sup>a</sup>* = \begin{bmatrix}
                -2 & 0 \\
                0 & -2
                \end{bmatrix}

1. b) |*H<sup>a</sup>*| = 4, so the matrix is negative definite, this means it is a concave function and it has a global maximum at x = 0 and y = 0.

```julia
Hᵃ = [-2 0; 0 -2]
```

1. c)

```julia
@assert det(Hᵃ) == 3
```

1. d) Eigenvalues

```julia
eigvals(Hᵃ)
```

1. e) Function to plot

```julia
fa(x,y) = -x^2 - y^2
```

```julia
plot(xrange, xrange, fa, st = :surface, title = "\$a=0\$")
```
